```python
# Recuento de descripciones de productos

most_frequent = df1[['Description', 'StockCode']].value_counts().reset_index()
most_frequent
```

```python
most_frequent[most_frequent['StockCode'] == '85123A']
```

La lógica es que para cada StockCode, elegimos la descripción más frecuente.

```python
# Agrupamos por 'StockCode' y nos quedamos con el registro más frecuente

most_frequent = most_frequent.groupby('StockCode').head(1)
most_frequent
```

```python
# Renombramos las columnas
most_frequent.columns = ['freq_Description', 'StockCode', 'count']

# Unimos los DataFrames
df2 = df1.merge(most_frequent, on='StockCode', how='left')
df2.head()
```

```python
df2['Description'] = df2['freq_Description']
df2.head()
```

```python
df2.isnull().sum()
```

Ahora vamos a eliminar los registros que quedan sin descripción.

```python
df2.dropna(subset=['Description'], inplace=True)
df2.isnull().sum()
```

Ahora en df2 vamos a eliminar las columnas extras que hemos añadido.

```python
df2.drop(columns=['freq_Description', 'count'], inplace=True)
df2.head()
```

```python
df2.describe()
```

En el campo de Quantity, UnitPrice, observamos que el valor mínimo es un número negativo alto.

```python
df2[df2['Quantity'] < 0]
```

Estas cantidad negativas podrían ser refunds de productos.

```python
df2[df2['UnitPrice'] < 0]
```

```python
df3 = df2[(df2['Quantity'] > 0) & (df2['UnitPrice'] > 0)]
df3.describe()
```

```python
df3['Quantity'].quantile(0.9999)
```

```python
df3[df3['Quantity'] > 1500]
```

Asumimos en este punto que los valores contenidos en el DataFrame son correctos.

### Feature Engineering: Creación de Nuevas Columnas

```python
df4 = df3.copy()
df4['TotalSales'] = df4['Quantity'] * df4['UnitPrice']
df4.head()
```

```python
df4['Month'] = df4['InvoiceDate'].dt.month
df4.head()
```

# Visualizaciones y EDA

## 1. Gráfico ventas mensuales

```python
monthly_sales = df4.groupby('Month')['TotalSales'].sum().reset_index()

fig = px.line(monthly_sales, x='Month', y='TotalSales',
              markers=True,
              title='Evolución de Ventas Mensuales',
              line_shape='spline')

fig.update_traces(
    line=dict(color='royalblue', width=3)  # Color y grosor de la línea
)

# Ajustar la apariencia de los ejes
fig.update_layout(
    xaxis_title="Mes",  # Etiqueta del eje X
    yaxis_title="Ventas Totales ($)",  # Etiqueta del eje Y
    xaxis=dict(showgrid=False),  # Ocultar líneas de la cuadrícula en X
    yaxis=dict(showgrid=True, gridcolor="lightgrey", tickprefix="$"),  # Mantener grid en Y con prefijo de moneda
    template="plotly_white"  # Fondo blanco y diseño limpio
)

fig.show()
```

**Insights**

La tendencia que nos muestra el gráfico es que claramente a finales de año, empezandoe en Agosto, las ventas aumentan drásticamente hasta tener un pico en Noviembre, esto puede ser provocado por la época de Navidad. Justo después de Navidad, se observa que las ventas disminuyen.

## 2. Top 5 países

```python
top_5_countries = df4.groupby('Country')['TotalSales'].sum().sort_values(ascending=False).head(5).reset_index()

fig = px.bar(top_5_countries, x='TotalSales', y='Country',
             orientation='h',
             title='Top 5 Países por Total Ventas',
             text='TotalSales',
             color='Country',
             color_discrete_sequence=px.colors.qualitative.Set2
             )

fig.update_layout(
    xaxis_title="Ventas Totales",
    yaxis_title="País",
    template="plotly_white",  # Estilo limpio
    yaxis=dict(categoryorder='total ascending')  # Ordenar las barras de menor a mayor
)

fig.show()
```

```python
# Calcular el total de ventas global
total_sales_global = df4['TotalSales'].sum()

# Calcular el porcentaje de contribución
top_5_countries = (df4.groupby('Country')['TotalSales']
                    .sum()
                    .sort_values(ascending=False)
                    .head(5)
                    .reset_index())

top_5_countries['Percentage'] = (top_5_countries['TotalSales'] / total_sales_global) * 100

# Crear el gráfico con porcentajes
fig = px.bar(top_5_countries,
             x='Percentage',  # Usamos el porcentaje en lugar de las ventas totales
             y='Country',
             orientation='h',
             title='Top 5 Países por Contribución (%) en Ventas Totales',
             text=top_5_countries['Percentage'].map(lambda x: f'{x:.2f}%'),  # Mostrar % en las barras
             color='Country',
             color_discrete_sequence=px.colors.qualitative.Set2
             )

# Personalizar diseño
fig.update_layout(
    xaxis_title="Contribución al Total de Ventas (%)",
    yaxis_title="País",
    template="plotly_white",
    yaxis=dict(categoryorder='total ascending')  # Ordenar las barras de menor a mayor
)

fig.show()
```

## 4. Análisis RFM

```python
# Supongamos que la fecha actual es la fecha de la última factura + 1 día
current_dt = df4['InvoiceDate'].max() + pd.Timedelta(days=1)
current_dt
```

```python
rfm = df4.groupby('CustomerID').agg({
    'InvoiceDate':lambda x: (current_dt - x.max()).days,
    'InvoiceNo':'count',
    'TotalSales':'sum'
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm.head()
```

```python
# Segmentación de clientes basado en RFM
rfm['R_Segment'] = pd.qcut(rfm['Recency'], 4, labels=[4, 3, 2, 1])
rfm['F_Segment'] = pd.qcut(rfm['Frequency'], 4, labels=[1, 2, 3, 4])
rfm['M_Segment'] = pd.qcut(rfm['Monetary'], 4, labels=[1, 2, 3, 4])
rfm['RFM_Score'] = rfm[['R_Segment', 'F_Segment', 'M_Segment']].sum(axis=1)
```

```python
rfm.sample(5)
```

```python
# Clientes con el RFM_score más alto
rfm.sort_values(by='RFM_Score', ascending=False)
```

## 5. Tasa de cancelación

```python
customer_last_purchase = df4.groupby('CustomerID')['InvoiceDate'].max()
customer_last_purchase.head()
```

```python
customer_last_purchase = (current_dt - customer_last_purchase).dt.days
customer_last_purchase.head()
```

```python
# Definimos el límite de la tasa de cancelación (por ejemplo, 90 días sin comprar)
churn_threshold = 90
churned_customers = customer_last_purchase[customer_last_purchase > churn_threshold]
churned_customers.head(5)
```

```python
print('Número de Churned Customers:', len(churned_customers))

fig = px.histogram(customer_last_purchase.to_pandas(), x='InvoiceDate',
                   title='Distribución de la pérdida de clientes',
                   color_discrete_sequence=['#1b4965']
                   )

# Añadir una línea vertical
fig.add_vline(x=churn_threshold, line=dict(color='black', dash='dash', width=2))

# Actualizar el diseño del gráfico
fig.update_layout(
    xaxis_title="Días desde la Última Compra",
    yaxis_title="Recuento de Clientes",
    template="plotly_white"  # Estilo limpio
)

fig.show()
```
